## What is a Function?
A **function** is a named, reusable sequence of statements that performs a specific computation. Think of a function as a "black box": you don't necessarily need to know the complex logic happening inside; you only need to know what to put in (input) and what you get out (output).

### Function Call
To execute a function, you **call** it by its name followed by parentheses. Inside the parentheses, we provide an **argument**.

````python
name = 'Melody'
type(name)
````

**Key Terms:**
*   **Function Name:** `type` (the instruction).
*   **Argument:** `Melody` (the data you pass into the function).
*   **Return Value:** `<str>` (the result the function "hands back" to you).

It is common to say that a function **"takes"** an argument and **"returns"** a result.

## Why Use Functions?
In professional software development, functions are the building blocks of clean code.

1.  **Readability & Abstraction:** Naming a group of statements (e.g., `calculate_tax()`) makes code self-documenting. You hide complex details behind a simple name.
2.  **DRY Principle (Don't Repeat Yourself):** If you need to perform the same task 10 times, you write the function once and call it 10 times.
3.  **Maintainability:** If a formula changes, you only have to update the code in one place (the function definition).
4.  **Modular Testing:** You can test individual functions to ensure they work before using them in a larger program.

## Best Practice: "Shadowing"
You must treat built-in function names as **reserved**. If you name a variable `max` or `len`, you "shadow" (override) the built-in function.

````python
# BAD PRACTICE
len = 5             # Now 'len' is an integer, not a function
print(len("Hello"))
````

**Tip:** In most code editors, built-in functions will be highlighted in a specific color. If your variable name turns that color, change the name!

## User-Defined Functions in Python

### Introduction
Up until now, we have been using functions written by other people (like `print`, `len`, or `math.sqrt`). In professional programming, the real power comes from **User-Defined Functions**: blocks of code you design, name, and reuse to solve your specific problems.

Think of a function as a **Recipe Card**:
*   **Defining the function:** Writing the instructions on the card and filing it away in a drawer. (The kitchen doesn't smell like food yet!)

## Function Definition
To define a function, we use a specific syntax. Let's look at a simple example:

````python
def greet():
    print("Hello, welcome to Python!")

greet()
````

- `def` - Keyword used to define a function
- `greet` - Function name
- `()` - Parentheses (can contain parameters)
- `:` - Marks the beginning of the function body
- `Indented code` - Statements that belong to the function

#### Function Parameters

`Parameters` are variables that receive values when a function is called.

````python
def greet(name):
    print(f"Hello, {name}")

greet("Melody")
````

- `name` is a parameter
- `Melody` is an argument

#### Functions with Multiple Parameters

````python
def add_numbers(a, b):
    print(a + b)

add_numbers(5, 3)
````

### Returning Values

Functions can send results back using the return keyword.

### Why Return Instead of Print?

Using return allows the result to be stored and reused.

````python
def square(x):
    return x ** 2

result = square(5)

print(result + 10)
````

### Functions Without Parameters

### Variable Scope
Local Variables

Variables created inside a function only exist inside that function.

````python
def example():
    x = 10
    print(x)

example()
````

Trying to access `x` outside the function will cause an error.

### Global Variables

Variables created outside functions can be accessed inside functions.

````python
message = "Hello"

def show_message():
    print(message)

show_message()
````


### 2. Positional vs. Keyword Arguments
#### a) Positional Arguments
Arguments are matched to parameters in order:


````python
def print_twice(bruce):
    print(bruce)
    print(bruce)

print_twice(bruce="Bruce Lee")
````

````python
def introduce(name, age):
    print(f"My name is {name} and I'm {age} years old.")

introduce("Alice", 30)  # name="Alice", age=30
````

### b) Keyword Arguments
- You can specify parameters by name (order doesn’t matter):

````python
introduce(age=25, name="Bob")
````

### 3. Default Parameter Values
Define default values for parameters; callers may omit them:

````python
def greet(name="Friend", greeting="Hello"):
    print(f"{greeting}, {name}!")

greet()                      # "Hello, Friend!"
greet("Melody")              # "Hello, Melody!"
greet("Bonareri", "Welcome") # "Welcome, Bonareri!"
````

### Example: Student Grade Calculator

````python
def calculate_grade(score):
    if score >= 80:
        return "A"
    elif score >= 70:
        return "B"
    elif score >= 60:
        return "C"
    elif score >= 50:
        return "D"
    else:
        return "F"

grade = calculate_grade(75)
print(grade)
````

## Fruitful vs. Void Functions in Python

### Fruitful Function
- **Returns a value** using the `return` keyword.
- The result can be stored in a variable or used in expressions.

````python
def add(a, b):
    return a + b

result = add(5, 3)

print(result)
````

### Why are they called fruitful?

They "produce" or "yield" a result that can be stored in a variable, used in calculations, or passed to another function.

````python
def square(x):
    return x ** 2

answer = square(4) + 10

print(answer)
````

### Void Functions

A void function performs an action but does not explicitly return a value.

````python
def greet(name):
    print(f"Hello, {name}")

greet("Melody")
````

#### What Happens if We Store the Result of a Void Function?

````python
def greet(name):
    print(f"Hello, {name}")

result = greet("Melody")

print(result)
````

### 4. Variable-Length Arguments: *args and **kwargs
a) *args (Positional Variable Arguments)
- Collect any extra positional arguments as a tuple

````python
def sum_all(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_all(1, 2, 3, 4, 5))  # 10
````

- args inside the function is a tuple of all passed positional arguments beyond those explicitly named.

b) **kwargs (Keyword Variable Arguments)
- Collect any extra keyword arguments as a dictionary:

````python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Melody", age=25, country="Kenya")
````

- ```kwargs``` inside the function is a dict mapping parameter names to values for any keywords not captured by named parameters.

## Lamda Functions

### Introduction
Normally, when we want to create a function, we use the `def` keyword, give it a name, and write a block of code. However, sometimes we need a **quick, one-time-use function** that is so simple we don't even want to give it a name.

In Python, these are called **Lambda Functions** (also known as **Anonymous Functions**).

### The Key Characteristics:
*   **Anonymous:** They have no name (unless you assign them to a variable).
*   **Single-Line:** They can only contain **one single expression**.
*   **Auto-Return:** You don't type the word `return`. They automatically return the result of the expression.


## Anatomy of a Lambda
To understand a lambda, let's compare it to a standard function.

````python
def square(x):
    return x ** 2

square(5)
````


**Lambda Way:**
```python
# keyword  parameter : expression
lambda     x         : x ** 2
```

### Syntax Breakdown:
1.  **`lambda`**: The keyword that tells Python "this is an anonymous function."
2.  **Arguments**: The inputs (like `x` or `a, b`).
3.  **The Colon `:`**: Separates the inputs from the logic.
4.  **Expression**: The actual math or logic to be performed.

### Examples

### Example 1: Basic Math
You can assign a lambda to a variable to give it a "temporary name."

````python
# A function that adds 10 to any number
add_ten = lambda x : x + 10

print(add_ten(5))  # Output: 15
````

### Example 2: Multiple Arguments
Just like regular functions, lambdas can take multiple inputs.

````python
# A function to calculate the area of a rectangle
calculate_area = lambda length, width : length * width

print(calculate_area(10, 5)) # Output: 50
````

### Example 3: Conditional Logic (The "Ternary" Operator)
Even though lambdas are only one line, we can use them to make quick decisions using an `if/else` statement written on a single line.

````python
# A function to check if a player is strong enough to enter a dungeon
check_strength = lambda power : "Enter" if power > 50 else "Too Weak"

print(check_strength(65)) # Output: Enter
print(check_strength(30)) # Output: Too Weak
````
