### Special Methods in Python (a.k.a. Dunder Methods)

Special methods in Python are built-in methods that start and end with double underscores (e.g., `__init__`, `__str__`).  
They’re also known as **dunder methods** (short for *double underscore*) or **magic methods**.

These methods define how objects behave with built-in Python operations such as printing, indexing, iteration, and arithmetic.

---

### Common Special Methods

### 1. Object Initialization and Representation

#### `__init__(self, ...)`
- **Purpose**: Initializes a newly created object.
- **Called when**: An object is created.


````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person_1 = Person("Alice", 27)
````

#### `__str__(self)` -  Human-Readable Description
- **Purpose:** Used when you print an object. It returns a string that’s easy to read and meant for users of your program.
- **Called when:** The str() function or print() is used on the object.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"{self.name}, Age: {self.age}"
    
person_1 = Person("Alice", 30)
str(person_1)
````

`__repr__(self)` - Developer-Facing Description
- Purpose: gives a technical string version of the object, helpful for debugging.
- Called when: The repr() function is used or in an interactive session.

````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name='{self.name}', age={self.age})"

person = Person("Alice", 30)
repr(person)
````

#### 2. Object Comparison and Arithmetic
`__eq__(self, other)` - Equality check (==)

- Purpose: defines how two objects of your class are compared using the == operator.
- Called when: == operator is used

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __eq__(self, other):
        return self.x == other.x and self.y == other.y
    
p1 = Point(2, 3)
p2 = Point(7, 3)
print(p1 == p2)
````

- p1 == p2 checks if the x and y values are the same.
- If `__eq__` was not defined, Python would say False even if values matched, unless they were the same object.

`__ne__(self, other)`
- Purpose: defines how two objects are compared using the != operator.
- Called when: != operator is used

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __ne__(self, other):
        return not (self.x == other.x and self.y == other.y)

p1 = Point(2, 3)
p2 = Point(4, 5)
print(p1 != p2)  # Output: True
````

- self.x and self.y refer to the coordinates of the current object.
- other.x and other.y refer to the coordinates of the object being compared to.

`__lt__(self, other)`
- Purpose: Less than comparison (<)
- Called when: < operator is used

````python
class Number:
    def __init__(self, value):
        self.value = value

    def __lt__(self, other):
        return self.value < other.value

n1 = Number(5)
n2 = Number(10)
print(n1 < n2)  # Output: True
````

`__add__(self, other)`
- Purpose: Addition (+)
- Called when: + operator is used

````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2
print(v3)  # Output: Vector(4, 6)
````

### 3. Container Methods
`__len__(self)`
- Purpose: define what should be returned when someone calls the len() function on your custom object.
- It allows you to control how length is calculated for objects you define using classes.
- Called when: len() function is used

````python
class Collection:
    def __init__(self, items):
        self.items = items

    def __len__(self):
        return len(self.items)

coll = Collection([1, 2, 3])
coll_2 = Collection(["a", "b", "c", "d"])
print(len(coll)) 
print(len(coll_2))
````

`__len__` allows your object to behave like a list (or string, or dictionary):

If your class doesn't have the `__len__` method, and you try to call len() on an object of that class, Python will raise a TypeError.



`__getitem__(self, key)`
- Purpose: allows indexing into your object — just like you do with lists, dictionaries, or strings.
- Called when: accessing elements of your custom object using square brackets: obj[index].

````python
class MyList:
    def __init__(self, elements):
        self.elements = elements

    def __getitem__(self, index):
        return self.elements[index]

my_list = MyList([10, 20, 30])
print(my_list[1])  # Output: 20
````

#### 4. Object Lifecycle Methods
`__del__(self)`

- Purpose: gets called when an object is about to be destroyed — typically when there are no more references to it.
- Called when: Object’s reference count drops to zero

````python
class Resource:
    def __init__(self, name):
        self.name = name
        print(f"Resource {self.name} created.")

    def __del__(self):
        print(f"Resource {self.name} destroyed.")

res = Resource("MyResource")
del res  # Output: Resource MyResource destroyed.
````

## Object-Oriented Programming Principles
OOP allows objects to interact with each other using four basic principles: encapsulation, inheritance, polymorphism, and abstraction. 

These four OOP principles enable objects to communicate and collaborate to create powerful applications.

## **Encapsulation**
Encapsulation is the practice of bundling **Data** (Attributes) and **Methods** (Behaviors) into a single unit (a Class) and restricting direct access to some of that object's components.

**The "Capsule" Analogy:**
Think of a medical capsule.
*   **The Medicine (Data):** The chemical powder inside that cures you.
*   **The Shell (Class):** The coating that holds the powder together and protects it from the outside environment until it's ready to be used.

**The Goal:**
1.  **Bundling:** Keeping related data together.
2.  **Data Hiding:** Preventing the outside world from accidentally breaking the internal workings of an object.


## **Use Case: Bank Account Security**
*Why do we need to hide data?*

Imagine a banking app where the `balance` variable is **Public** (open to everyone).
```python
# DANGEROUS CODE
account.balance = 0  # Anyone can accidentally wipe out the savings!
```
If the variable is public, there is no control. A junior developer or a bug in another part of the code could accidentally set a negative balance or wipe the account.

**The Encapsulation Solution:**
We make the `balance` **Private**. We only allow changes through a specific method, like `withdraw()`, which contains security checks (authentication).

---

## **Access Modifiers**
Unlike Java or C++, Python does not use keywords like `public` or `private`. Instead, Python uses **Underscores (`_`)** to dictate access levels.

| Access Level | Syntax | Symbol | Meaning |
| :--- | :--- | :--- | :--- |
| **Public** | `self.name` | No Underscore | Open to everyone. |
| **Protected** | `self._project` | **One** Underscore | "Internal Use Only" (Convention). |
| **Private** | `self.__salary` | **Two** Underscores | "Do Not Touch" (Name Mangling). |

---

### **A. Public Members**
*   **Accessible:** Everywhere (Inside the class, outside the class).
*   **Industry Standard:** Use this for data that is safe for anyone to read or change (e.g., a User's first name).

````python
class Employee:
    def __init__(self, name, salary):
        #Public Attribute 
        self.name = name
        self.salary = salary

    def show_details(self):
        print(f"Name: {self.name}, Salary: {self.salary}")

emp1 = Employee("Annette", 60000)
print(emp1.name)
print(emp1.show_details())
````

### **B. Protected Members**
*   **Accessible:** Inside the class and its **Subclasses** (Child classes).
*   **Syntax:** Prefix with a **single underscore** `_`.
*   **Crucial Note for Python:**
    *   Python does *not* strictly enforce this. You *can* technically access `emp._project` from the outside without an error.
    *   **However:** In the professional industry, seeing a `_` means **"Warning: Do not touch this variable directly unless you are writing a subclass."** It is a signal to other developers.

````python
# class Company:
#     def __init__(self):
#         #protected member
#         self._project = "Python project"

# Child class
class Employee():
    def __init__(self, name):
        self.name = name
        self._project = "Python project"

        super().__init__()

    def show_project(self):
        # Accessing the protected member
        print(f"{self.name} is working on: {self._project}")

emp1 = Employee("Mark")
emp1.show_project()
````


### **C. Private Members**
*   **Accessible:** **ONLY** inside the class where they are defined.
*   **Syntax:** Prefix with **double underscores** `__`.
*   **Mechanism:** Python performs **Name Mangling**. It internally changes the variable name so it cannot be found easily from the outside.

````python
class Employee():
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary # private member


    def get_salary(self):
        # Accessing the private member
        print(f"Salary is: {self.__salary}")

emp1 = Employee("Mark", 10000)
emp1.get_salary()
````

## **Comprehensive Example: The "Safe" Employee System**
*This example combines all three levels to show a realistic class structure.*

````python
class Employee:
    def __init__(self, name, id, salary, project):
        self.name = name
        self.id = id
        self._project = project # protected member
        self.__salary = salary #private member


    #Getter method
    def get_salary(self):
        return self.__salary

    #Setter method
    def set_salary(self, new_amount):
        if new_amount > 0:
            self.__salary = new_amount
            print(f"Salary updated")
        else:
            print("Error: Salary cannot be negative")

emp1 = Employee('James', 100, 20000, "Data Analysis")
emp1.get_salary()
emp1.set_salary(1000)
````

## **How to Access Private Members (Getters & Setters)**
If `__salary` is private, how do we update it? We use **Public Methods** to control the access. This acts as a "Gatekeeper."

````python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary

    # GETTER: Allows read-only access
    def get_salary(self):
        return self.__salary

    # SETTER: Allows modification with Validation Logic
    def set_salary(self, new_amount):
        if new_amount > 0:
            self.__salary = new_amount
            print("Salary updated.")
        else:
            print("Error: Salary cannot be negative.")

emp = Employee("James", 5000)


# We cannot do: emp.__salary = -1000
# We MUST use the method:
emp.set_salary(-1000) # Output: Error: Salary cannot be negative.
emp.set_salary(6000)  # Output: Salary updated.
````

## **Abstraction**
Abstraction is the principle of **hiding the complex implementation details** and showing only the essential features of an object.

It allows the user to interact with a system using a clean, simple **Interface** without needing to understand the complicated logic happening in the background.

*   **The "What" vs. The "How":** Abstraction focuses on **what** an object does, not **how** it does it.

### **Real-World Analogy: The Car Dashboard**
*   **The Interface:** When you drive, you use a steering wheel, a gas pedal, and a brake.
*   **The Abstraction:** You do not need to know how the fuel injection system works, how the combustion engine fires, or how the brake pads apply friction. All that complexity is **hidden**.
*   **The Result:** You can drive a Tesla, a Ford, or a Suzuki because they all adhere to the same "Interface" (Steering wheel + Pedals), even though they work completely differently under the hood.

## **Why is Abstraction Important?**
*   **Simplicity:** It reduces complexity by hiding unnecessary details.
*   **Maintainability:** You can change the internal code (the engine) without breaking the external interface (the steering wheel).
*   **Security:** It hides critical internal logic from the outside world.
*   **Enforced Structure:** It forces all developers to follow a strict structural plan (e.g., "Every Car MUST have a mileage method").


## **How to Achieve Abstraction in Python**
Unlike Java, Python does not have built-in abstract classes by default. We must import the **`abc` module** (Abstract Base Classes).

### **The Three Rules of Abstraction**
1.  **Inheritance:** You must inherit from `ABC`.
2.  **Decorator:** You must use `@abstractmethod` on methods you want to enforce.
3.  **No Instantiation:** You **cannot** create an object from an Abstract Class. If you try `c = Car()`, Python will throw an error.


````python
from abc import ABC, abstractmethod

class Car(ABC):
    
    @abstractmethod
    def mileage(self):
        pass

class Tesla(Car):
    def mileage(self):
        print("The mileage is 30kmph (Electric)")

class Suzuki(Car):
    def mileage(self):
        print("The mileage is 25kmph (Petrol)")


tesla_1 = Tesla()
suzuki_1 = Suzuki()

tesla_1.mileage()
suzuki_1.mileage()
````

**Key Takeaway:** The `Car` class defined the **Rule** (everyone must have mileage), but the subclasses defined the **Logic** (what that mileage actually is).

## **Example 2: The Geometric Shapes (Polygons)**
*Here, we define a structural contract: All Polygons must be able to tell us how many sides they have.*

````python
from abc import ABC, abstractmethod

# 1. The Abstract Base Class
class Polygon(ABC):
    
    @abstractmethod
    def sides(self):
        """Every polygon must define how many sides it has."""
        pass

# 2. Concrete Classes
class Triangle(Polygon):
    def sides(self):
        print("Triangle has 3 sides")

class Pentagon(Polygon):
    def sides(self):
        print("Pentagon has 5 sides")

class Hexagon(Polygon):
    def sides(self):
        print("Hexagon has 6 sides")

class Square(Polygon): # Note: Class names should be PascalCase
    def sides(self):
        print("Square has 4 sides")

# --- EXECUTION ---
# Creating objects
t = Triangle()
s = Square()
p = Pentagon()
k = Hexagon()

# Invoking the common interface
t.sides() # Output: Triangle has 3 sides
s.sides() # Output: Square has 4 sides
p.sides() # Output: Pentagon has 5 sides
k.sides() # Output: Hexagon has 6 sides
````

### Types of Polymorphism in Python

#### 1. **Polymorphism in Action (Duck Typing)**
In many languages (like Java), classes must be related (via Inheritance) to use polymorphism.
**Python is different.** Python follows a philosophy called **Duck Typing**:
*"If it walks like a duck and quacks like a duck, then it must be a duck."*

This means: **Python doesn't care about the Class Name.** It only cares: **"Does this object have the method I am trying to call?"**

Even though Dog and Cat are different types, Python doesn't care — as long as they both have a speak() method.

````python
class Dog:
    def speak(self):
        return "Bark!"
    
class Cat:
    def speak(self):
        return "Meow!"
    
dog = Dog()
cat = Cat()

print(dog.speak())
print(cat.speak())
````

````python
class Tiger:
    def nature(self):
        print("I am a tiger")

class Lion:
    def nature(self):
        print("I am a Lion")


tiger = Tiger() 
lion = Lion()


tiger.nature()
lion.nature()
````

### 2. Method Overriding (Polymorphism with Inheritance)
When a child class redefines a method from its parent class.

````python
class Animal:
    def speak(self):
        return "Some sound"


class Dog(Animal):
   def speak(self):
        return "Bark!"
   
class Cat(Animal):
    def speak(self):
        return "Meow!"
    
dog = Dog()
cat = Cat()

print(dog.speak())
print(cat.speak())
````

````python
class Bird:
    def intro(self):
        print("There are many types of birds.")

    def flight(self):
        print("Most birds can fly but some cannot.")

class Sparrow(Bird):
    def flight(self):
        print("Sparrows can fly.")

class Ostrich(Bird):
    def flight(self):
        print("Ostriches cannot fly.")

# Polymorphic Function
def activate_bird(bird_obj):
    # This function works for ANY bird
    bird_obj.intro()
    bird_obj.flight()

# Usage
s = Sparrow()
o = Ostrich()

activate_bird(s) # Output: Sparrows can fly.
activate_bird(o) # Output: Ostriches cannot fly.
````

## **Inheritance**
Inheritance is a fundamental mechanism in Object-Oriented Programming that allows a new class to acquire the properties (attributes) and behaviors (methods) of an existing class.

It models an **"Is-A" Relationship**.
*   A `Dog` **is an** `Animal`.
*   A `Car` **is a** `Vehicle`.

### **Why do we use it?**
1.  **Code Reusability (DRY Principle):** "Don't Repeat Yourself." If you have logic that applies to all Animals (like `eat()` or `sleep()`), you write it once in the `Animal` class. All subclasses (`Dog`, `Cat`, `Lion`) get it for free.
2.  **Extensibility:** You can modify the behavior of a specific child class without breaking the code in the parent class.
3.  **Organization:** It creates a hierarchical structure that mirrors the real world.

### **Terminology**
*   **Parent Class** (also called **Base** or **Super** Class): The class being inherited from.
*   **Child Class** (also called **Derived** or **Sub** Class): The class that inherits.
---
## Types of Inheritance
1.  Single Inheritance
2.  Multiple Inheritance
3.  Multilevel Inheritance

### **Single Inheritance**
**Definition:** Occurs when a derived class inherits from **only one** base class. This is the simplest and most common form of inheritance.

**How it works:** The Child class automatically has access to every public attribute and method defined in the Parent.

````python
class Parent:
    def greet(self):
        print("Hello from the Parent class!")

class Child(Parent):
    pass

c = Child()
c.greet()  # Inherited from Parent
````

````python
# 1. The Base Class (Parent)
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return f"{self.name} makes a sound."

# 2. The Derived Class (Child)
# Syntax: class ChildName(ParentName):
class Dog(Animal):
    # We override the speak method to be specific to dogs
    def speak(self):
        return f"{self.name} says Woof!"

# --- EXECUTION ---
# Creating an instance of the Child class
dog = Dog("Buddy")

# 1. Accessing data inherited from Animal
print(dog.name)   # Output: Buddy

# 2. Calling the method defined in Dog
print(dog.speak()) # Output: Buddy says Woof!
````

**Detailed Explanation:**
*   When we created `dog = Dog("Buddy")`, Python looked for an `__init__` in `Dog`. It didn't find one!
*   So, it automatically went up to the `Animal` class and used *that* `__init__`. This is why `dog` has a `name` attribute even though we didn't write it inside the `Dog` class.

### **Multiple Inheritance**
**Definition:** Occurs when a derived class inherits from **more than one** base class.

**Use Case:** This is often used for "Mixins"—classes that provide specific skills (like `Flyable` or `Swimmable`) rather than a full identity.

````python
# Base Class 1
class Flyer:
    def fly(self):
        return "Flying high!"

# Base Class 2
class Swimmer:
    def swim(self):
        return "Swimming in the water!"

# Derived Class (Inherits from BOTH)
class Duck(Flyer, Swimmer):
    def quack(self):
        return "Quacking!"

# --- EXECUTION ---
duck = Duck()

# The duck object has skills from all three classes
print(duck.fly())   # Output: Flying high! (From Parent 1)
print(duck.swim())  # Output: Swimming in the water! (From Parent 2)
print(duck.quack()) # Output: Quacking! (From Child)
````

**Technical Note (Method Resolution Order - MRO):**
If both `Flyer` and `Swimmer` had a method with the *same name* (e.g., `move()`), Python needs to decide which one to use. It follows the order you list them in the parentheses.
*   `class Duck(Flyer, Swimmer)`: Python checks `Flyer` first.
*   `class Duck(Swimmer, Flyer)`: Python checks `Swimmer` first.

### **Multilevel Inheritance**
**Definition:** Occurs when a class is derived from another derived class, creating a "Family Tree" or chain of inheritance.
*   **Structure:** Grandparent -> Parent -> Child.

````python
# Level 1: Grandparent
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return f"{self.name} makes a sound."

# Level 2: Parent (Inherits from Animal)
class Mammal(Animal):
    def walk(self):
        return f"{self.name} is walking."

# Level 3: Child (Inherits from Mammal)
class Dog(Mammal):
    def speak(self):
        return f"{self.name} says Woof!"
    
# --- EXECUTION ---
dog = Dog("Buddy")

# 1. Method from Dog (Immediate class)
print(dog.speak()) # Output: Buddy says Woof!

# 2. Method from Mammal (Parent)
print(dog.walk())  # Output: Buddy is walking.

# 3. Attribute from Animal (Grandparent)
print(dog.name)    # Output: Buddy
````

## **Polymorphism & Method Overriding**
Inheritance is not just about copying code; it's about **specializing** it.

**What is Method Overriding?**
Method Overriding is the process where a child class provides a specific implementation of a method that is already defined in its parent class.
*   **Rule:** The method name and parameters must be the same.
*   **Result:** When you call the method on the child object, Python runs the child's version, effectively "ignoring" the parent's version for that object.

````python
class Animal:
    def sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def sound(self):
        print("Bark")

class Cat(Animal):
    def sound(self):
        print("Meow")

# Using the classes
d = Dog()
c = Cat()

d.sound()  # Bark
c.sound()  # Meow
````

**Comprehensive Example: The Vehicle System**

````python
# The Generic Parent Class
class Vehicle:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price

    def show(self):
        print('Details:', self.brand, self.model, 'Price:', self.price)

    def max_speed(self):
        print('Vehicle max speed is 160')

    def gear_system(self):
        print('Vehicle has 6 shifter gearbox')

# The Specific Child Class
class Car(Vehicle):
    # OVERRIDING: A Car is faster than a generic vehicle
    def max_speed(self):
        print('Car max speed is 260')

    # OVERRIDING: A Car works differently than a generic vehicle
    def gear_system(self):
        print('Car has Automatic Transmission')

# --- EXECUTION & COMPARISON ---

# 1. The Car Object (The Child)
print("--- Car Object ---")
car = Car('Audi', 'R8', 9000000)

car.show()          
# Works via Inheritance (Car didn't define it, so it uses Vehicle's version)

car.max_speed()     
# Works via Overriding (Uses Car's version) -> Output: Car max speed is 260

car.gear_system()   
# Works via Overriding (Uses Car's version) -> Output: Car has Automatic Transmission

# 2. The Vehicle Object (The Parent)
print("\n--- Vehicle Object ---")
vehicle = Vehicle('Nissan', 'Magnite', 550000)

vehicle.show()        # Uses Vehicle's version
vehicle.max_speed()   # Uses Vehicle's version -> Output: Vehicle max speed is 160
vehicle.gear_system() # Uses Vehicle's version -> Output: Vehicle has 6 shifter gearbox
````

**Extending Parent Functionality (`super`)**

Sometimes, you don't want to completely **replace** the parent's method (Override). You want to **keep** the parent's behavior and just **add** a little bit extra to it.

We use the **`super()`** function to call the Parent method from inside the Child.

````python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager(Employee):
    def __init__(self, name, salary, department):
        # 1. Let the Parent handle the standard stuff (name/salary)
        super().__init__(name, salary)
        
        # 2. Now handle the Manager-specific stuff
        self.department = department

mgr = Manager("Alice", 90000, "IT")
print(mgr.department) # IT
print(mgr.name)       # Alice (Handled by super!)
````

### The @property Decorator 
- The @property decorator in Python is a powerful feature that allows you to control access to instance variables in a class. 
- It provides a way to implement getters, setters, and deleters while keeping the syntax clean and intuitive.

### Basics of the @property Decorator

- The @property decorator turns a method into a “getter” for an attribute, allowing you to access an instance variable as if it were a regular attribute, but with the option to add custom logic when getting, setting, or deleting the value.

````python
class Car:
    def __init__(self, speed):
        self._speed = speed

    @property
    def speed(self):
        """Getter method for the speed attribute."""
        print("Getting the speed")
        return self._speed

    @speed.setter
    def speed(self, value):
        """Setter method for the speed attribute."""
        if value < 0:
            raise ValueError("Speed cannot be negative")
        print("Setting the speed")
        self._speed = value

    @speed.deleter
    def speed(self):
        """Deleter method for the speed attribute."""
        print("Deleting the speed")
        del self._speed


car = Car(50)        # Creating a Car instance with an initial speed of 50
print(car.speed)     # Getting the speed (using the getter method)
car.speed = 100      # Setting the speed to 100 (using the setter method)
del car.speed        # Deleting the speed attribute (using the deleter method)
````

### `__init__` Method
- Initializes the `_speed` attribute.
- The underscore (`_`) is a naming convention to indicate the attribute is private (internal use only).

### `@property` (Getter)
- Allows access to a method like an attribute.
- Called when you do `car.speed`.
- Makes the class interface cleaner (you don’t need `get_speed()`).

### `@speed.setter` (Setter)
- Called when assigning a value like `car.speed = 100`.
- You can add validation logic, e.g., check that the speed is not negative.

### `@speed.deleter` (Deleter)
- Called when you delete the attribute with `del car.speed`.
- Useful for cleanup or logging.

### Why Use `@property`?
- **Encapsulation**: Controls how an attribute is accessed and modified.
- **Ease of Use**: Allows you to use simple dot notation (`car.speed`) while still having logic behind it.
- **Backward Compatibility**: You can start with a normal attribute, then add logic later using `@property` without changing how external code interacts with the class.
