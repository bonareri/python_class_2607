## What is a Function?
A **function** is a named, reusable sequence of statements that performs a specific computation. Think of a function as a "black box": you don't necessarily need to know the complex logic happening inside; you only need to know what to put in (input) and what you get out (output).

### Function Call
To execute a function, you **call** it by its name followed by parentheses. Inside the parentheses, we provide an **argument**.

````python
name = 'Melody'
type(name)
````

**Key Terms:**
*   **Function Name:** `type` (the instruction).
*   **Argument:** `Melody` (the data you pass into the function).
*   **Return Value:** `<str>` (the result the function "hands back" to you).

It is common to say that a function **"takes"** an argument and **"returns"** a result.

## Why Use Functions?
In professional software development, functions are the building blocks of clean code.

1.  **Readability & Abstraction:** Naming a group of statements (e.g., `calculate_tax()`) makes code self-documenting. You hide complex details behind a simple name.
2.  **DRY Principle (Don't Repeat Yourself):** If you need to perform the same task 10 times, you write the function once and call it 10 times.
3.  **Maintainability:** If a formula changes, you only have to update the code in one place (the function definition).
4.  **Modular Testing:** You can test individual functions to ensure they work before using them in a larger program.

### 4. Variable-Length Arguments: *args and **kwargs
a) *args (Positional Variable Arguments)
- Collect any extra positional arguments as a tuple

````python
def sum_all(*args):
    total = 0
    for num in args:
        total += num
    return total

print(sum_all(1, 2, 3, 4, 5))  # 10
````

- args inside the function is a tuple of all passed positional arguments beyond those explicitly named.

b) **kwargs (Keyword Variable Arguments)
- Collect any extra keyword arguments as a dictionary:

````python
def print_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_info(name="Melody", age=25, country="Kenya")
````

- ```kwargs``` inside the function is a dict mapping parameter names to values for any keywords not captured by named parameters.

## Lamda Functions

### Introduction
Normally, when we want to create a function, we use the `def` keyword, give it a name, and write a block of code. However, sometimes we need a **quick, one-time-use function** that is so simple we don't even want to give it a name.

In Python, these are called **Lambda Functions** (also known as **Anonymous Functions**).

### The Key Characteristics:
*   **Anonymous:** They have no name (unless you assign them to a variable).
*   **Single-Line:** They can only contain **one single expression**.
*   **Auto-Return:** You don't type the word `return`. They automatically return the result of the expression.


## Anatomy of a Lambda
To understand a lambda, let's compare it to a standard function.

````python
def square(x):
    return x ** 2

square(5)
````

````python
square = lambda x: x * x

print(square(5))
````


**Lambda Way:**
```python
# keyword  parameter : expression
lambda     x         : x ** 2
```

### Syntax Breakdown:
1.  **`lambda`**: The keyword that tells Python "this is an anonymous function."
2.  **Arguments**: The inputs (like `x` or `a, b`).
3.  **The Colon `:`**: Separates the inputs from the logic.
4.  **Expression**: The actual math or logic to be performed.

### Examples

### Example 1: Basic Math
You can assign a lambda to a variable to give it a "temporary name."

````python
# A function that adds 10 to any number
add_ten = lambda x : x + 10

print(add_ten(5))  # Output: 15
````

### Example 2: Multiple Arguments
Just like regular functions, lambdas can take multiple inputs.

````python
# A function to calculate the area of a rectangle
calculate_area = lambda length, width : length * width

print(calculate_area(10, 5)) # Output: 50
````

### Example 3: Conditional Logic (The "Ternary" Operator)
Even though lambdas are only one line, we can use them to make quick decisions using an `if/else` statement written on a single line.

````python
# A function to check if a player is strong enough to enter a dungeon
check_strength = lambda power : "Enter" if power > 50 else "Too Weak"

print(check_strength(65)) # Output: Enter
print(check_strength(30)) # Output: Too Weak
````

## Recursive Functions

### The Concept: The Search for the Key
Imagine you are looking for a **Golden Key** that is hidden inside a large wooden box. When you open the wooden box, you don't find a key—instead, you find **another, smaller box**. 

To find that key, you have to:
1.  Open the box you just found.
2.  If it contains a smaller box, **repeat the exact same process.**
3.  If you find the **Golden Key**, you stop! (This is the **Base Case**).

## Introduction
**Recursion** is a programming technique where a function solves a big problem by calling a "mini version" of itself to solve a smaller part of that same problem. 
A **recursive function** is a function that **calls itself** in order to solve a problem by breaking it down into **smaller, simpler subproblems**. 
Recursion is commonly used when a problem naturally fits a repeated or self-similar structure.


## The Two Pillars of a Recursive Function
For a recursive function to work without crashing your computer, it must have two specific parts:

### A. The Base Case (The "Stop" Sign)
This is the condition that tells the function to stop calling itself. In our analogy, the Base Case is **finding the key**. Without a base case, you would keep looking for boxes forever, even after the key is found.

### B. The Recursive Case (The "Next Step")
This is where the function calls itself again, but with a **simpler or smaller** input. Each time the function calls itself, it must get one step closer to the Base Case.


## How it Works: The Call Stack
When a function calls itself, Python keeps track of each call in a structure called the **Call Stack**. Each call waits for the next one to finish before it can complete its own work. It "pauses" and waits for the mini-version to give it an answer.

Think of it like **Delegating a Task**:
1.  **Boss:** "Tell me the total of 5 + 4 + 3 + 2 + 1."
2.  The Boss is busy, so they ask a **Manager**: "Hey, tell me the sum of numbers up to 4, and I will just add 5 to your answer."
3.  The Manager asks an **Employee**: "Tell me the sum up to 3, and I'll add 4 to it."
4.  This continues until someone is asked for the sum of **1**. That person simply says **"1"** (The Base Case).
5.  Now, the answers travel back up the chain until the Boss gets the final total.

### Example 1: The Factorial

The factorial of `4` (written as `4!`) is `4 * 3 * 2 * 1`. 

**The Logic:**
To find `4!`, you calculate `4 * 3!`.
To find `3!`, you calculate `3 * 2!`.
...until you reach `1!`, which is just `1`.

````python
def factorial(n):
    # 1. Base Case: If n is 1, we finally have an answer!
    if n == 1:
        return 1
    
    # 2. Recursive Case: n * (n - 1)
    else:
        # The function PAUSES here until factorial(n-1) returns a value
        return n * factorial(n - 1)

print(factorial(4)) # Output: 24
````

### Example 2: Fibonacci Numbers
The Fibonacci sequence starts `0, 1, 1, 2, 3, 5, 8...` where every number is the sum of the two before it.

````python
def fibonacci(n):
    # Base Case: The first two positions (0 and 1) are just themselves
    if n <= 1:
        return n
    
    # Recursive Case: "To find this number, find the two before it and add them"
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(6)) # Output: 8
````

### Recursion vs. Loops
You might ask: *"Why not just use a `for` loop?"*

*   **Loops (Iteration)** are generally faster and use less memory. They are better for simple tasks like counting to 100.
*   **Recursion** is better for "branching" problems—tasks where one step leads to two or more new steps (like searching through all the folders and sub-folders on your computer).

### The Danger: "Stack Overflow"
If you forget your Base Case, Python will keep creating "Waiting Rooms" until it runs out of memory. You will see an error called: **RecursionError: maximum recursion depth exceeded.**


### Challenge: The "Counting Down" Function
**The Task:** Write a recursive function called `count_down(n)`. 
1.  It should print the number `n`.
2.  If `n` is greater than 0, it should call itself with `n - 1`.
3.  If `n` is 0, it should print "Blastoff!" and stop.

````python
def count_down(n):
    if n == 0:
        print("Blastoff!") # Base Case
    else:
        print(n)
        count_down(n - 1) # Recursive Case

count_down(5)
````

## Try Except

### Introduction to Exceptions
Even if your Python syntax is perfect, your program can still crash. These "runtime errors" are called **Exceptions**.

**The Problem:**
Think about the `int()` function. It works fine if the user types `"20"`, but if they type `"twenty"`, the program crashes with a `ValueError`. In a real app, you don't want the whole program to shut down; you want to tell the user, "Hey, please enter a number!"

````python
# This will crash the program if the user types a word
age = int(input("Enter your age: "))
print(f"You are {age} years old.")
````

### The Solution: The `try / except` Block
Python provides a way to "catch" these errors before they crash the script. Think of `try` as a **Safety Net**.

### How it Works:
1.  **`try`**: Python "attempts" to run the code inside this block.
2.  **`except`**: If an error occurs inside the `try` block, Python **jumps immediately** to the `except` block instead of crashing.

````python
try:
    age = int(input("Enter your age: "))
    print(f"You are {age} years old.")
except:
    print("Error: Please enter a valid whole number.")

print("The program is still running!")
````

````python
inp = input("Enter Farenheit temperature: ")
farenheit = float(inp)
celsius = (farenheit - 32) * 5 / 9
print("Temperature in Celsius:", celsius)
````

````python
inp = input("Enter Farenheit temperature: ")
try:
    farenheit = float(inp)
    celsius = (farenheit - 32) * 5 / 9
    print("Temperature in Celsius:", celsius)
except:
    print("Invalid input. Please enter a valid number.")
````

## Handling Specific Exceptions
A "catch-all" `except` block is okay, but it is better to handle specific types of errors. This allows you to give the user better feedback.

### Common Exception Types:
*   **`ValueError`**: When a function gets an argument of the right type but inappropriate value (like `int("hello")`).
*   **`ZeroDivisionError`**: When you try to divide a number by zero.
*   **`TypeError`**: When an operation is applied to an object of the wrong type.

````python
try:
    number = int(input("Enter a number to divide 100 by: "))
    result = 100 / number
    print(f"Result: {result}")
except ValueError:
    print("Invalid Input: You must enter a number.")
except ZeroDivisionError:
    print("Math Error: You cannot divide by zero.")
````

## Exceptions Inside Functions
The best way to use `try/except` is inside a function. This makes your function **"Robust"** (meaning it can handle bad data without breaking the rest of the app).

````python
def safe_divide(a, b):
    try:
        ans = a / b
        return ans
    except ZeroDivisionError:
        return "Error: Division by zero is not allowed."

# Calling the function
print(safe_divide(10, 2)) # Output: 5.0
print(safe_divide(10, 0)) # Output: Error: Division by zero is not allowed.
````

## Example: The Smart Travel Estimator

## The Real-World Scenario
When you use apps like Uber, Google Maps, or DoorDash, the "Estimated Time of Arrival" (ETA) isn't just a simple calculation. The app has to account for unpredictable factors like distance and weather.

In this example, we will build a script that calculates travel time while accounting for random weather conditions.

### The Goal
Our program must:
1. **Randomize the Journey:** Pick a random distance for the trip.
2. **Simulate the Environment:** Randomly pick a weather condition (Clear, Rainy, or Snowy).
3. **Apply Logic:** Use conditionals to decide how much the weather will slow us down.
4. **Provide a "Safe Estimate":** Use the `math` module to ensure the final time is a whole number, rounding up to ensure the user isn't late.

## The Logic
Before we code, we need to define our rules:
* **Clear Skies:** Normal speed (1.0x multiplier).
* **Heavy Rain:** 50% slower (1.5x multiplier).
* **Snow Storm:** 200% slower (3.0x multiplier).

## The Code Implementation

````python
import random
import math

print("--- Smart Route Estimator ---")

# 1. THE JOURNEY: Generate a random distance between 5 and 50 miles
distance = random.randint(5, 50)
speed = 60 # We assume a base speed of 60 miles per hour (1 mile per minute)

# 2. THE CONDITION: Pick a random weather condition from a list
weather = random.choice(["Clear", "Rainy", "Snowy"])

# 3. THE MULTIPLIER: Use conditionals to set the delay based on weather
if weather == "Snowy":
    multiplier = 3.0
elif weather == "Rainy":
    multiplier = 1.5
else:
    multiplier = 1.0

# 4. THE CALCULATION: 
# Time = (Distance / Speed) * 60 minutes * multiplier
raw_time = (distance / speed) * 60 * multiplier

# 5. THE POLISH: Use math.ceil
# We round UP because it is better to tell a user they will arrive in 
# 15 minutes and have them arrive early, rather than telling them 14 
# and having them arrive late.
final_time = math.ceil(raw_time)

# 6. THE OUTPUT: Display the results to the user
print("Distance to destination: " + str(distance) + " miles")
print("Current Weather: " + weather)
print("Estimated Travel Time: " + str(final_time) + " minutes")
````
