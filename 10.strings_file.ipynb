## STRINGS IN PYTHON
### What is a String?
A string is a data type in Python used to represent text. It's a sequence of characters enclosed in:

- Single quotes ' '
- Double quotes " "
- Triple quotes ''' ''' or """ """ (for multi-line text)

````python
single = 'Hello'
double = "World"
triple = '"This is a triple quoted string"'
````

### 1. String Indexing
- `String indexing` allows you to `access individual characters` in a string using their position (index).
- In Python, strings are ordered, meaning each character has a fixed position.
- Indexing starts at 0 (not 1)

````python
text = "Python"
````

#### Accessing Characters Using Indexing

Use square brackets `[]` with the index number.

````python
text[2]
````

#### String Indexing Rules
- Strings are immutable (cannot be changed)
- You can read characters, but you cannot modify them
- Trying to access an index that does not exist causes an error.
> `This will cause an error`:

````python
text
````

````python
text[0] = 'J'  # This will raise an error because strings are immutable
````

````python
text[7]
````

### 2. String Slicing
- `String slicing` allows you to `extract a portion (substring)` from a string using a range of indices.
- Instead of accessing one character, slicing returns multiple characters.
- Slicing Syntax:
```python
string[start : stop : step]
```
- `start` - index to begin (inclusive)
- `stop` - index to end (exclusive)
- `step` - interval between characters (optional)

````python
s = 'programming'
````

### Basic slicing

````python
s[2:5]
````

````python
s
````

#### Omitting the start and stop

````python
s[:3]
````

#### Using step in slicing

````python
s[0:11:2]
````

````python
s[::2]
````

````python
s
````

````python
s = "programmings"
````

````python
s[0:11]
````

### Negative indexing

````python
s[::-1]
````

### 3. Common String Methods
a. Case Methods

````python
text = "Hello World"
text.lower()
text.upper()
text.title()
````

b. Whitespace Removal

````python
msg = "   clean me   "
msg.strip()
msg.lstrip()
msg.rstrip()
````

c. Replacement & Searching

````python
sentence = "Data is power"
new_sentence = sentence.replace("power", "knowledge")
print(new_sentence)
print(sentence)
````

e. Count & Find

````python
phrase = "hello hello hello"
print(phrase.count("hello"))
print(phrase.find("llo"))
````

- `.count()` counts how many times a substring appears in the string.
- `.find()` searches for the first occurrence of the substring "llo" in the string.
- It returns the index (position) of the first character where the substring starts.

### Looping through Strings

Looping through a string means accessing `each character` one by one and performing an action on it.

Since strings are `iterable`, you can loop through them easily in Python.

#### Using a `for` Loop

````python
fruit = "banana"
for i in fruit:
    print(i)
````

#### Looping Using Indexing

You can loop through a string using its index positions.

````python
word = "python"
# len(word)
for i in range(len(word)):
    print(word[i])
````

#### Using `while` loop

````python
word = "python"
i = 0
while i < len(word):
    print(word[i])
    i += 1
````

### Looping Through a String with Conditions
#### Example: Print only vowels

````python
word = "Education"
for char in word:
    if char.lower() in 'aeiou':
        print(char)
````

## FILE HANDLING IN PYTHON
In Python, file handling lets us create, read, update, and delete files. This is useful for saving data permanently — unlike variables, which are lost once the program ends.

### In-Built File Handling Functions

| Function      | Purpose                                    |
| ------------- | ------------------------------------------ |
| `open()`      | Opens a file and returns a file object     |
| `write()`     | Writes text to the file                    |
| `read()`      | Reads the entire file content as a string  |
| `readline()`  | Reads a single line                        |
| `readlines()` | Reads all lines and returns them in a list |
| `close()`     | Closes the file                            |
| `with`        | Automatically manages opening and closing  |

### What is mode in Python File Handling?
When you open a file in Python using the open() function, you must specify a file mode. The mode tells Python what you want to do with the file:

- Do you want to read from it?
- Do you want to write to it?
- Do you want to add content to the end?
- Are you working with text or binary data?

syntax
```open("filename", "mode")```


| Mode  | Description                         |
|-------|-------------------------------------|
| `'r'` | Read (default)                      |
| `'w'` | Write (overwrites existing file)    |
| `'a'` | Append (adds to the end of file)    |
| `'x'` | Create new file (fails if exists)   |
| `'b'` | Binary mode (used for non-text files) |


#### 1. open() – Opening a File

````python
file = open('example_2.txt', 'w')
````

#### 2. write() – Writing to a File

````python
file.write("This is my first line.\n")
file.write("This is my second line.\n")
file.close()
````

#### 3. read() – Reading Entire File Content

````python
file = open('example.txt', 'r')
content = file.read()
print(content)
file.close()
````

#### 4. readline() – Read a Single Line

````python
file = open('example.txt', 'r')
line1 = file.readline()
file.close()
print(line1)
````

#### 5. readlines() – Read All Lines into a List

- Reads all lines and returns them as a list of strings
- Useful when you want to loop over the lines later

````python
file = open('example.txt', 'r')
lines = file.readlines()
file.close()
print(lines)
````

#### 6. close() – Closing a File

- Closes the file after you’re done with it
- Very important to free system resources and avoid file corruption

#### 7. Using with – Automatically Handles close()

- Best practice for working with files
- Automatically calls file.close() for you

````python
with open('example.txt', 'r') as file:
    content = file.read()
    print(content)
````

### File Modes in Action

####  1. w – Write Mode

````python
# creates and overwite the file
with open('anthem.txt', 'w') as file:
    file.write("O God of all creation,\n")
    file.write("Bless this our land and nation.\n")
````

#### 2. a – Append Mode

````python
# Add more content to the file
with open('anthem.txt', 'a') as file:
    file.write("Justice be our shield and defender,\n")
    file.write("May we dwell in unity.\n")
````

#### 3. x – Exclusive Creation Mode

````python
# Tries to create a new file, but errors if it already exists
with open('anthem2.txt', 'x') as file:
    file.write("O God of all creation,\n")
    file.write("Bless this our land and nation.\n")
````

#### 4. r – Read Mode

````python
with open('anthem.txt', 'r') as file:
    content = file.read()
    print(content)
````

### What is Binary Mode in File Handling?
In Python, files can be opened in Text Mode (default) or Binary Mode.

- Text Mode ('t'): Reads/writes strings (human-readable)
- Binary Mode ('b'): Reads/writes bytes (machine-readable)

Binary mode is useful when dealing with:

- Images (.jpg, .png)
- Audio files (.mp3)
- Video files (.mp4)
- PDFs, spreadsheets, etc.
- Any non-text file

Binary mode flags

| Mode   | Meaning                                       |
| ------ | --------------------------------------------- |
| `'rb'` | Read a binary file                            |
| `'wb'` | Write to a binary file (overwrites existing)  |
| `'ab'` | Append to a binary file                       |
| `'xb'` | Create a new binary file (error if it exists) |


````python
# open the original image in binary mode
with open("original_image.png","rb") as original:
    content = original.read()
    print(content[:20])
````

This reads raw bytes of the image — necessary for uploading to a server or sending over a network.

````python
with open("copied_image.png","wb") as copy:
    copy.write(content)
````
