## Lists

A **list** is a built-in Python data structure used to store **multiple items** in a **single variable**.  
Lists are:

- **Ordered**: Elements maintain their insertion order.
- **Mutable**: Can be changed (items can be added, removed, or modified).
- **Indexed**: Accessed by zero-based index.
- **Allow duplicates**: Multiple identical values are allowed.
- **Can hold mixed data types**: e.g., integers, strings, even other lists.

### List Comprehension

A `list comprehension` is a compact way to create a new list.

Think of it as a shortcut for a for loop that builds a list.

For example, suppose you have a list of numbers and want a new list containing their squares.

#### Traditional way

````python
numbers = [1, 2, 3, 4, 5]

squares = []

for num in numbers:
    squares.append(num ** 2)

print(squares)
````

#### List comprehension way

````python
numbers = [1, 2, 3, 4, 5]

squares = [num ** 2 for num in numbers]

print(squares)
````

Same result, but fewer lines.

Suppose we have a list of numbers and we want to:

1. Loop through each number.
2. Check if it’s even or odd.
3. Multiply even numbers by 2, odd numbers by 3.
4. Store the results in a new list.

````python
numbers = [1, 4, 7, 10, 13, 16]
processed_numbers = []

for i in numbers:
    if i % 2 == 0:
        new_num = i * 2
        print(f"{i} is even, multiplied by 2 = {new_num}")
    else:
        new_num = i * 3
        print(f"{i} is odd, multiplied by 3 = {new_num}")
    processed_numbers.append(new_num)

print("Original numbers:", numbers)
print("Processed numbers:", processed_numbers)
````

#### A function that returns a list of the numbers of the Fibonacci series

````python
def fib2(n):  # return Fibonacci series up to n
    """Return a list containing the Fibonacci series up to n."""
    result = []
    a, b = 0, 1
    while a < n:
        result.append(a)    # see below
        a, b = b, a+b
    return result

f100 = fib2(100)    # call it
f100 
````
